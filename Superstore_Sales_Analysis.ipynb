```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_management = pd.read_csv("Sample - Superstore.csv",encoding='latin1')
```

```python
sales_management
```

```python
sales_management.info()
```

```python
sales_management.describe()
```

```python
sales_management.drop('Country',axis=1, inplace=True)
```

```python
sales_management.duplicated().sum()
```

```python
sales_management.drop('Postal Code',axis=1, inplace=True)
```

```python
sales_management.isnull().sum()
```

```python
sales_management.head(3)
```

```python
sales_management['Order Date'] = pd.to_datetime(sales_management['Order Date'])
sales_management['Ship Date'] = pd.to_datetime(sales_management['Ship Date'])
```

```python
sales_management.info()
```

```python
sales_management['Order Date'].min()
```

```python
sales_management['Order Date'].max()
```

```python
yearly_profit=sales_management.groupby(
    sales_management['Order Date'].dt.year
)['Profit'].sum().sort_values(ascending=False)
yearly_profit
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    yearly_profit.index,
    yearly_profit.values,
    marker='o'
)

plt.title('Yearly Profit Trend')
plt.xlabel('Year')
plt.ylabel('Profit')

plt.grid(True)

plt.show()
```

```python
yearly_sales_profit = sales_management.groupby(
    sales_management['Order Date'].dt.year
)[['Sales','Profit']].sum()
yearly_sales_profit
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    yearly_sales_profit.index,
    yearly_sales_profit['Sales'],
    marker='o',
    label='Sales'
)

plt.plot(
    yearly_sales_profit.index,
    yearly_sales_profit['Profit'],
    marker='o',
    label='Profit'
)

plt.title('Yearly Sales and Profit Trend')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)

plt.show()
```

```python
monthly_sales_profit=sales_management.groupby(
    [sales_management['Order Date'].dt.year,
    sales_management['Order Date'].dt.month
    ])[['Sales','Profit']].sum()
monthly_sales_profit.index = [
    f"{year}-{month:02d}"
    for year, month in monthly_sales_profit.index
]
monthly_sales_profit
```

```python
plt.figure(figsize=(14,5))

plt.plot(
    monthly_sales_profit.index,
    monthly_sales_profit['Sales'],
    marker='o',
    label='Sales'
)

plt.plot(
    monthly_sales_profit.index,
    monthly_sales_profit['Profit'],
    marker='o',
    label='Profit'
)

plt.title('Monthly Sales and Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=60)
plt.legend()
plt.grid(True)

plt.show()
```

```python
sales_management.groupby('Product Name')[['Sales','Profit']].sum().sort_values(by=['Sales','Profit'],ascending=[False,True])
```

```python
top_products=sales_management.groupby('Product Name')[['Sales','Profit']].sum().sort_values(by=['Sales','Profit'],
    ascending=[False,True]).head(10)
top_products
```

```python
top_products[['Sales','Profit']].plot(
    kind='barh',
    figsize=(12,5)
)

plt.title('Top 10 Products by Sales vs Profit')
plt.xlabel('Amount')
plt.ylabel('Product Name')
plt.xticks(rotation=70)

plt.show()
```

```python
category_by_sales_profit=sales_management.groupby('Category')[['Sales','Profit']].sum().sort_values(
    by=['Sales','Profit'],ascending=False)
category_by_sales_profit
```

```python
category_by_sales_profit[['Sales','Profit']].plot(
    kind='barh',
    figsize=(12,5)
)
plt.title('Category by sales and profit')
plt.xlabel('Amount')
plt.ylabel('Category')
plt.xticks(rotation=0)

plt.show()
```

```python
sales_management.groupby(['Category','Sub-Category'])[['Profit','Sales']].sum().sort_values(by=['Sales','Profit'],ascending=[False,True])
```

```python
sales_management.groupby('Sub-Category').agg({
    'Discount': 'mean',
    'Profit' : 'sum'
}).sort_values(by=['Discount','Profit'] , ascending=[False,True])
```

```python
sales_management.groupby(['Category','Sub-Category']).agg({
    'Discount' : 'mean',
    'Profit' : 'sum'
}).sort_values(by='Discount' , ascending=False)
```

```python
sales_by_region=sales_management.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales' , ascending=False)
sales_by_region
```

```python
sales_by_region[['Sales','Profit']].plot(
    kind='bar',
    figsize=(12,5)
)

plt.title('Region by Sales and Profit')
plt.xlabel('Region Name')
plt.ylabel('Amount')
plt.xticks(rotation=0)

plt.show()
```

```python
sales_by_region=sales_management.groupby('Region')['Sales'].sum()
sales_by_region
```

```python
sales_by_region.plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title('Regional Contribution to Total Sales')
plt.axis('equal')
plt.show()
```

```python
sales_management.groupby('Segment')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
```
